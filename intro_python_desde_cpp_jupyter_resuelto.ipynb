## 1. Python vs C++

### En C++
- Debes declarar tipos explícitamente.
- Usas llaves `{}` para bloques.
- El punto y coma `;` es obligatorio.
- La compilación suele ser previa a la ejecución.

### En Python
- El tipo se infiere automáticamente.
- La indentación define los bloques.
- No se usan `;` al final.
- El código suele ejecutarse directamente.

### Ejemplo conceptual

**C++**
```cpp
int x = 5;
if (x > 0) {
    cout << "Positivo" << endl;
}
```

**Python**
```python
x = 5
if x > 0:
    print("Positivo")
```

## 5. Operadores básicos

Los operadores matemáticos son muy parecidos a C++:
- `+` suma
- `-` resta
- `*` multiplicación
- `/` división real
- `//` división entera
- `%` módulo
- `**` potencia

````python
a = 10
b = 3

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)
````

## 6. Condicionales

En C++:
```cpp
if (nota >= 7) {
    cout << "Aprueba";
} else {
    cout << "Reprueba";
}
```

En Python:

````python
nota = 8

if nota >= 7:
    print("Aprueba")
else:
    print("Reprueba")
````

````python
nota= 6

if nota >= 7:
    print("Aprueba")
else:
    print("Reprueba")
````

También existe `elif`, equivalente al `else if` de C++.

````python
temperatura = 30

if temperatura > 30:
    print("Hace calor")
elif temperatura >= 15:
    print("Clima templado")
else:
    print("Hace frío")
````

## 7. Bucles

### `for` en Python
Python recorre secuencias de forma muy natural usando `range()`.

````python
for i in range(11):
    print("i =", i)
````

### Comparación rápida con C++

**C++**
```cpp
for (int i = 0; i < 5; i++) {
    cout << i << endl;
}
```

**Python**
```python
for i in range(5):
    print(i)
```

### `while` en Python

````python
contador = 0

while contador < 3:
    print("contador =", contador)
    contador += 1
````

## 8. Listas: una estructura muy usada

En C++ podrías pensar en algo parecido a `vector`.
En Python, una lista es dinámica y muy flexible.

````python
numeros = [10, 20, 30, 40]

print(numeros)
print("Primer elemento:", numeros[0])
print("Último elemento:", numeros[len(numeros)-1])
print("Último elemento version 2:", numeros[-1])
````

````python

len(numeros) # Porque len cuenta cuantos valores, pero al estar indexando desde cero
# el utimo valor es len -1
````

````python
numeros.append(50) #agrega valores
print(numeros)

numeros.remove(20) #Quita el valor que esta como parametro
print(numeros)
````

## 10. Funciones

## 14. Primer contacto con programación orientada a objetos

Ya que vienes de C++, esto te resultará familiar, pero la sintaxis es más limpia.

````python
class Estudiante:
    def __init__(self, nombre, nota):
        self.nombre = nombre
        self.nota = nota

    def mostrar_info(self):
        print(f"Estudiante: {self.nombre}, Nota: {self.nota}")

est1 = Estudiante("María", 9.5)
est1.mostrar_info()
````

## 16. Mini ejercicio guiado

Crea una lista de 5 números y muestra:
- la suma
- el promedio
- el valor máximo

````python
numeros = [5, 8, 12, 20, 7]

suma = sum(numeros)
promedio = suma / len(numeros)
maximo = max(numeros)

print("Lista:", numeros)
print("Suma:", suma)
print("Promedio:", promedio)
print("Máximo:", maximo)
````
